```python
# import packages
from scipy import *
from numpy import *
from matplotlib.pyplot import *
from math import e, sqrt

eps = 0.1

t = arange(0., 1., 0.001)

while eps > 10**-4:
    a = -1 + sqrt(1 - eps)
    b = -1 - sqrt(1 - eps)

    c1 = (1 + sqrt(1 - eps))/(2*sqrt(1 - eps))
    c2 = (-1 + sqrt(1 - eps))/(2*sqrt(1 - eps))
    
    u = c1*e**(a*t) + c2*e**(b*t)
    
    plot(t, u, label='$\epsilon = $' + str(eps))
    
    eps = eps/10
    
plot(t, ones(len(t)), label='$\epsilon = 0$')
legend(loc = 3)
figsize(10, 6)
show()
```

```python
# import packages
from scipy import *
from numpy import *
from matplotlib.pyplot import *
from math import e, sqrt
from numpy.linalg import norm
from scipy.integrate import ode
```

```python
class Body:
    def __init__(self, rgPosInit, rgVelInit, dMassInit):
        self.rgPos = rgPosInit
        self.rgVel = rgVelInit
        self.dMass = dMassInit
    def getState(self):
        return [self.rgPos, self.rgVel]
    def setState(self, rgPosSet, rgVelSet):
        self.rgPos = rgPosSet
        self.rgVel = rgVelSet
```

```python
class System:
    def __init__(self, rgBodiesInit = None):
        if rgBodiesInit is None:
            rgBodiesInit = []
        self.rgBodies = rgBodiesInit
    def getSize(self):
        return len(self.rgBodies)
    def getBody(self, iBody):
        if iBody >= len(self.rgBodies) or -iBody > len(self.rgBodies):
            print 'ERROR: index out of range'
            return
        return self.rgBodies[iBody]
    def setBody(self, iBody, BodySet):
        if iBody >= len(self.rgBodies) or -iBody > len(self.rgBodies):
            print 'ERROR: index out of range'
            return
        self.rgBodies[iBody] = BodySet
    def addBody(self, BodyAdd, iBody = -1):
        self.rgBodies.insert(iBody, BodyAdd)
    def rmBody(self, iBody = -1):
        return self.rgBodies.pop(iBody)
    def getSysState(self):
        rgSysState = []
        for iBody in self.rgBodies:
            rgSysState.append(iBody.getState())
        return rgSysState
    def setSysState(self, rgSysState):
        for iState in len(rgSysState):
            self.rgBodies.setState(iState[0], iState[1])
    def __str__(self):
        strSys = 'System: \n'
        for iBody in range(len(self.rgBodies)):
            strSys = strSys + 'Body ' + str(iBody) + ':\n'
            strSys = strSys + '\tLoc: ' + str(self.rgBodies[iBody].rgPos) + '\n'
            strSys = strSys + '\tVel: ' + str(self.rgBodies[iBody].rgVel) + '\n'
            strSys = strSys + '\tMass: ' + str(self.rgBodies[iBody].dMass) + '\n'
        return strSys
```

```python
class Force:
    def __init__(self, SystemInit, rgtInit):
        self.System = SystemInit
        self.rgt = rgtInit
    def __call__(self, System, rgt):
        rgSysState = self.System.getSysState()
        self.rgDeriv = []
        self.rgAcc = zeros((len(rgSysState), len(rgSysState)))
        for iState in range(len(rgSysState)):
            self.rgDeriv.append([rgSysState[iState][1], 0])
            for jState in range(len(rgSysState)):
                if jState >= iState:
                    self.rgAcc[iState, jState, :] = zeros([len(System.rgBodies[iState]), 1)
                else:
                    self.rgAcc[iState, jState, :] = self.fnAcc(System.rgBodies[iState], \
                                                    System.rgBodies[jState])
        
        for iState in len(rgSysState):
            self.rgDeriv[iState][1] = self.System.rgBodies[iState].dMass * \
                (sum(self.rgAcc[:, iState, :]) - sum(self.rgAcc[iState, :, :]))
        
        return self.rgDeriv
             
    # calculates the force of Body1 on Body 2
    def fnAcc(Body1, Body2):
        return Body1.dMass * Body2.dMass * (Body2.rgPos - Body1.rgPos) \
                / norm(Body2.rgPos - Body1.rgPos)**3
```

```python
Body1 = Body(array([0., 0.]), array([0., 0.]), 1.)
Body2 = Body(array([1., 0.]), array([0., 0.8]), 1.)
Body3 = Body(array([-1., 0.]), array([0., -0.8]), 1.)

Sys = System([Body1, Body2, Body3])

f = Force(Sys, [])

print f.System
rgSysState = f.System.getSysState()
print rgSysState
print rgSysState[1][1]
```
